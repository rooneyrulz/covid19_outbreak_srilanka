```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
covid19_confirmed_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
# covid19_confirmed_df
```

```python
df_confirmed_sl = covid19_confirmed_df.loc[covid19_confirmed_df['Country/Region'] == 'Sri Lanka']
```

```python
covid19_death_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
# covid19_death_df
```

```python
df_death_sl = covid19_death_df.loc[covid19_death_df['Country/Region'] == 'Sri Lanka']
```

```python
covid19_recovered_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
# covid19_recovered_df
```

```python
df_recovered_sl = covid19_recovered_df.loc[covid19_recovered_df['Country/Region'] == 'Sri Lanka']
```

```python
plt.figure(figsize=(8, 5), dpi=100)

plt.bar(df_confirmed_sl.columns[4:], df_confirmed_sl.iloc[0, 4:].values, label='Confirmed Cases', color='blue')
plt.bar(df_recovered_sl.columns[4:], df_recovered_sl.iloc[0, 4:].values, label='Recovered Cases', color='green')
plt.bar(df_death_sl.columns[4:], df_death_sl.iloc[0, 4:].values, label='Deaths', color='red')


# plt.plot(df_confirmed_sl.columns[4:], df_confirmed_sl.iloc[0, 4:].values, label='Confirmed Cases', linewidth=2, color='blue', markersize=10, marker='.', markeredgecolor='blue')
# plt.plot(df_recovered_sl.columns[4:], df_recovered_sl.iloc[0, 4:].values, label='Recovered Cases', linewidth=2, color='green', markersize=10, marker='.', markeredgecolor='green')
# plt.plot(df_death_sl.columns[4:], df_death_sl.iloc[0, 4:].values, label='Deaths', color='red', linewidth=2, markersize=10, marker='.', markeredgecolor='red')

plt.title('COVID-19 Outbreak in Sri Lanka')
plt.xlabel('date')
plt.ylabel('cases')
plt.xticks(covid19_confirmed_df.columns[8::8])

plt.legend()
plt.grid()

plt.show()
```
